### iterate through the folders to create a dataframe of the form filepaths labels

```python
dataset_path = '/Users/nirwanaamir/Downloads/datasetskripsi'
data = []
labels = []
```

### split df into train_df, test_df and valid_df

```python
# Loop melalui setiap kategori makanan
for category in os.listdir(dataset_path):
    category_path = os.path.join(dataset_path, category)
    
    # Skip non-directory files (e.g., .DS_Store)
    if not os.path.isdir(category_path):
        continue

    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)
        
        # Membaca gambar dengan benar
        img_array = convert_image_to_array(img_path)
        if img_array is not None:
            data.append(img_array)
            labels.append(category)
```

### create train, test, valid  generators

```python
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)
```

### create function to show some image examples

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
def augment_data(X_train):
    aug = ImageDataGenerator(
        rotation_range=25, width_shift_range=0.1, height_shift_range=0.1,
        shear_range=0.2, zoom_range=0.2, horizontal_flip=True, fill_mode="nearest"
    )

    # Menambahkan dimensi tambahan pada data
    X_train_augmented = np.expand_dims(X_train, axis=-1)

    aug.fit(X_train_augmented)
    return aug
```

### create the model

```python
# Membangun model CNN
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dense(len(set(labels)), activation="softmax"))
```

```python
# Compile model
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Melatih model dengan augmented data
model.fit(aug.flow(X_train, y_train, batch_size=32), epochs=10, validation_data=(X_test, y_test))
```

```python
accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy[1]*100:.2f}%")
```
