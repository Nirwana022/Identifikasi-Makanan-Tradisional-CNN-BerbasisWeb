```python
import cv2
import matplotlib.pyplot as plt

# Function to resize image
def resize_image(image_path, target_height, target_width):
    # Read image
    image = cv2.imread(image_path)
    # Resize image
    resized_image = cv2.resize(image, (target_width, target_height))
    return resized_image

# Path to the image
image_path = "/Users/nirwanaamir/Downloads/dataset-2/Barongko (Sulawesi Selatan)/3.jpeg"  # Ganti dengan path gambar Anda

# Parameters
height = 224
width = 224

# Resize image
resized_image = resize_image(image_path, height, width)

# Display original and resized images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Resized image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Inisialisasi ImageDataGenerator dengan parameter augmentasi
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=30,
    horizontal_flip=True
)

# Path ke gambar yang akan di-augmentasi
img_path = '/Users/nirwanaamir/Downloads/dataset-2/Barongko (Sulawesi Selatan)/3.jpeg'

# Load gambar
img = load_img(img_path)

# Konversi gambar ke array
x = img_to_array(img)

# Reshape array menjadi bentuk (1, *shape) karena `flow` memerlukan array 4 dimensi
x = x.reshape((1,) + x.shape)

# Flow generator
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(batch[0])
    i += 1
    if i % 4 == 0:  # Mengehentikan augmentasi setelah 4 gambar yang di-generate
        break

plt.show()
```
